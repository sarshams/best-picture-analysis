```python
import pandas as pd
import sqlite3
import json
import seaborn as sns
```

```python
query = """SELECT id, running_time_in_minutes, title, year, year_ceremony, release_info.date
        FROM movies
        LEFT JOIN release_info ON movies.id == release_info.movie_id"""
```

```python
def get_config_params(src_dir ='./assets/', config_file='config.json'):
    with open(src_dir + config_file) as f:
        return json.load(f)
config_params = get_config_params()
db_filepath = config_params['database path']

data = pd.read_sql(query, sqlite3.connect(db_filepath))
```

```python
data['date'] = pd.to_datetime(data['date'])
data['month'] = data['date'].dt.month
```

```python
ax = sns.countplot(x='month', data=data)
```
